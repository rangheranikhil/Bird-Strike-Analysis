Number of people injured: No. of people injured during the incidents

Is Aircraft large : whether the aircraft is large or not

```python
# Checking duplicates and null values
df.duplicated().sum()  # 0
df.isnull().sum()
```

```python
# Deleting null values
df.dropna(subset=['Aircraft: Type'], inplace=True)
df.dropna(subset=['Aircraft: Number of engines?','Origin State'], inplace=True)
df.drop(columns = ['Remarks'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.info() # 4.6+ mb
```

```python
# changing dtype to category
cols = ['Aircraft: Type', 'Altitude bin', 'Wildlife: Number struck', 'Effect: Indicated Damage', 'Wildlife: Size', 'Conditions: Sky',
        'Pilot warned of birds or wildlife?', 'Is Aircraft Large?']

df[cols] = df[cols].astype('category')
```

```python
# removing unwanted char from string and change dtype to int
df['Cost: Total $'] = df['Cost: Total $'].str.replace(',', '').astype('int64')
df['Feet above ground']  = df['Feet above ground'].str.replace(',', '').astype('int64')
```

```python
# removing row and changing dtype to int8
df.drop(6664, inplace=True)
df['Aircraft: Number of engines?'] = df['Aircraft: Number of engines?'].astype('int8')
```

```python
# Exporting Data
df.to_csv('Bird Strike Cleaned.csv', index=False)
```
